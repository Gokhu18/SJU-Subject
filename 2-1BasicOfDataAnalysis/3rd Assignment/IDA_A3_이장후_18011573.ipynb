```python
df_ready_pca = df_apriori.iloc[:,1:np.size(df_apriori.iloc[0,:])]
#PCA를 통해 변수들 간의 상관관계를 다시 한 번 알아볼 수 있도록 한다.
#모든 변수를 one hot 인코딩된 Categorical data 로 놓고 진행하는 것이기 때문에, 딱히 정규화를 하지 않아도 된다.
from sklearn.decomposition import PCA


#df_pca_num = PCA(n_components=(df_ready_pca.columns.size-1))
df_pca_num = PCA(n_components=10)
df_pca_num_res = df_pca_num.fit(df_ready_pca)


PCA_result = ['Variance', 'Variance\%','Cum\%']
df_pca_num_res_pc = np.transpose(np.array(df_pca_num_res.components_).reshape(np.shape(df_pca_num_res.components_)))
df_pca_num_res_var = np.array(np.transpose(df_pca_num_res.explained_variance_[:,np.newaxis]))
df_pca_num_res_varp = np.array(np.transpose(df_pca_num_res.explained_variance_ratio_[:,np.newaxis]))
df_pca_num_res_sum = np.cumsum(np.round(df_pca_num_res.explained_variance_ratio_, decimals=8)*100)
df_pca_num_res_varc = np.array(np.transpose(df_pca_num_res_sum[:,np.newaxis]))
df_pca_num_res_tab = DataFrame(np.concatenate((df_pca_num_res_pc, df_pca_num_res_var, 
                                                   df_pca_num_res_varp, df_pca_num_res_varc), 
                                                  axis = 0),
                                   index = np.concatenate((np.array(df_ready_pca.columns), PCA_result)))

df_pca_num_res_tab
```

신기하게도, PC9까지만 사용해도 전체 변동의 85% 이상이 나온다. 하지만 데이터를 들여다 보면 Male과 Female 은 명확한 음의 상관관계를 가질수밖에 없기 때문에, 차원이 획기적으로 줄어들수밖에 없다. PC가 10 개라는것은 결국 맨 처음 데이터보다 조금 더 많은 Column의 양이다. Column 이 너무 많다 보니까, 이 모든 PC들에 어떤 변수가 어떻게 녹아있는지는 알아보기 힘들지만 모델에 쓰일 정도는 될 것 같아 보인다
